Advent of Code 2025 Day 3

```python
with open('input.txt','r') as f:
    data = f.read()
data = data.split('\n')[:-1]
```

Part 1

```python
total = 0
for d in data:
    digit=str(max(d[:-1]))
    digit+=str(max(d[d.index(max(d[:-1]))+1:]))
    total+=int(digit)
print(f'Total joltage: {total}')
        
```

Part 2

```python
total = 0
for d in data:
    digit=''
    for i in range(11,-1,-1):
        test = max(d[:len(d)-i])
        digit+=str(test)
        d=d[d.index(test)+1:]
    total+=int(digit)
print(f'Total joltage: {total}')
        
```
