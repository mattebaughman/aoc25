Advent of Code 2025 Day 4

Part 1

```python
import numpy as np
```

```python
with open('input.txt','r') as f:
    data = f.read()
data=data.split('\n')[:-1]
```

```python
def generate_coords_list(i,j):
    coords_list=[]
    for a in [-1,0,1]:
        for b in [-1,0,1]:
            if [a,b]!=[0,0]:
                coords_list.append([i+a,j+b])
    coords_list=[k for k in coords_list if (k[0]<140 and k[0]>=0 and k[1]<140 and k[1]>=0)]
    return(coords_list)
```

```python
tot=0
for i in range(len(data)):
    for j in range(len(data[i])):
        if data[i][j]=='@':
            coords=generate_coords_list(i,j)
            count=0
            for c in coords:
                if data[c[0]][c[1]]!='.':
                    count+=1
            if count<4:
                tot+=1
print(f'Number of rolls: {tot}')
```

Part 2

```python
big_tot = 0
data = np.array([np.array(list(d)) for d in data])
tot=1
while tot>0:
    tot=0
    for i in range(len(data)):
        for j in range(len(data[i])):
            if data[i,j]=='@':
                coords=generate_coords_list(i,j)
                count=0
                for c in coords:
                    if data[c[0],c[1]]!='.':
                        count+=1
                if count<4:
                    tot+=1
                    data[i,j]='.'
    big_tot+=tot
print(f'Number of rolls: {big_tot}')
```
