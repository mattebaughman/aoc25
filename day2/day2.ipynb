Part 1

```python
with open('input.txt','r') as f:
    data = f.read()
data=data.split('\n')[0]
data=data.split(',')
data=[k.split('-') for k in data]
```

```python
total=0
for k in data:
    for i in range(int(k[0]),int(k[1])+1):
        i = str(i)
        if len(i)%2==0:
            if i[:int(len(i)/2)]==i[int(len(i)/2):]:
                total+=int(i)
print(f'Sum of twice repeated numbers: {total}')
```

Part 2

```python
total=0
for k in data:
    for i in range(int(k[0]),int(k[1])+1):
        i = str(i)
        for min_factor in range(2,len(i)+1):
            if len(i)%min_factor==0:
                break
        for sub_length in range(1,int(len(i)/min_factor)+1):
            if len(i)%int(sub_length)==0:
                if i[:sub_length]*int(len(i)/sub_length)==i:
                    total+=int(i)
                    break
print(f'Sum of repeated numbers: {total}')
```
