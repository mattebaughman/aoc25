Advent of Code: Day 1

Part 1

```python
start_value = 50
```

```python
with open('input.txt','r') as f:
    instructions = f.read()
instructions=instructions.split('\n')[:-1]
```

```python
instructions=[-int(k[1:]) if k[0]=='L' else int(k[1:]) for k in instructions]
```

```python
current_value = start_value
total_zeroes = 0
for k in range(len(instructions)):
    current_value+=instructions[k]
    if current_value%100==0:
        total_zeroes+=1
print(f'Total zeroes (aka "password"): {total_zeroes}')
```

Part 2

```python
current_value = start_value
total_zeroes = 0
for k in range(len(instructions)):
    for i in range(abs(instructions[k])):
        current_value+=instructions[k]/abs(instructions[k])
        if current_value%100==0:
            total_zeroes+=1
print(f'Total zeroes (aka "password"): {total_zeroes}')
```
