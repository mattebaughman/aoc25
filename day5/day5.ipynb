Part 1

```python
with open('input.txt','r') as f:
    data = f.read()

data = data.split('\n\n')
data[0]=data[0].split('\n')
data[1]=data[1].split('\n')[:-1]
```

```python
fresh = [k.split('-') for k in data[0]]
ids = data[1]
total = 0
for i in ids:
    is_fresh=False
    for f in fresh:
        if int(f[0])<=int(i)<=int(f[1]):
            is_fresh=True
    if is_fresh:
        total+=1
print(total)
```

Part 2

```python
import pandas as pd
```

```python
fresh = pd.DataFrame(fresh).astype('int')
fresh.sort_values(by=0,inplace=True)
fresh=fresh.values
```

```python
total = 0
current=0
for f in fresh:
    if f[1]>current:
        if f[0]>current:
            total+=(f[1]-f[0]+1)
        else:
            total+=(f[1]-current)
        current=f[1]
print(total)
```
